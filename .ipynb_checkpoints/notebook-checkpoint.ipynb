```python
endpoint = 'https://fapi.binance.com/fapi/v1/klines'
res = requests.get(endpoint + '?symbol=BTCUSDT&interval=1d') #15m
dataArr = json.loads(res.content)
data = [None] * len(dataArr)
for k,v in enumerate(dataArr):
    data[k] = { 'timestamp': (int(v[0]) / 1000), 'close': float(v[3]) }

df = pd.DataFrame(data)
#df = df.set_index('timestamp')
#df.index = pd.to_datetime(df.index, unit='s')

df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
df.set_index('timestamp', inplace=True)

# hist = pd.DataFrame(data)
# hist = hist.set_index('time_col')
# hist.index = pd.to_datetime(hist.index, unit='s')
# target_col = 'close'

#df = pd.read_json(res.content)
# df = pd.json_normalize(dataArr)
df.head()

```

```python
roll_mean = df.rolling(window=8, center=False).mean()
roll_std = df.rolling(window=8, center=False).std()


fig = plt.figure(figsize=(16,6))
plt.plot(df, label='Original')
plt.plot(roll_mean, color='red', label='Rolling Mean')
plt.plot(roll_std, color='black', label = 'Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

### The Dickey-Fuller Test

```python
from statsmodels.tsa.stattools import adfuller

dftest = adfuller(df)

# Extract and display test results in a user friendly manner
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dftest)
```

```python
#The p-value is .7(1) here, so the time series is not stationary!
print ('Results of Dickey-Fuller test: \n')

print(dfoutput)
```

```python
# Drop non full weeks
df = df.iloc[5:-5 , :]
```

```python
# Use pandas grouper to group values using annual frequency
weeks_groups = df.groupby(pd.Grouper(freq='1W'))
#display(df.head(10))
#weeks_groups.head()
weeks_groups.head()
```

```python
# Create a new DataFrame and store yearly values in columns 
hist_weekly = pd.DataFrame()
#print(len(weeks_groups))
for idx, group in weeks_groups:
    #print(len(group))
    hist_weekly[idx.week] = group.values.ravel()

# Plot the yearly groups as subplots
#hist_weekly.plot(figsize=(16,16), subplots=True, legend=True);

hist_weekly.plot(figsize=(16,6), legend=True);
```
